###Data Preprocessing & Cleaning

```python
df_refined = df.copy()
df_refined
```

```python
df_refined['sex'] = df_refined['sex'].replace({'male' : 0, 'female' : 1})
df_refined.head()
```

```python
df_refined['smoker'] = df_refined['smoker'].replace({'yes' : 1, 'no' : 0})
df_refined.head()
```

```python
df_refined.rename(columns = {
    'sex' : 'is_female',
    'smoker' : 'is_smoker'
}, inplace = True)

df_refined.head()
```

```python
# One Hot Enoding

df_refined = pd.get_dummies(df_refined, columns=['region'], dtype='int64', drop_first=-True)
```

```python
df_refined.head()
```

###Feature Engineering & Extraction

```python
df_refined['bmi_category'] = pd.cut(
    df_refined['bmi'],
    bins = [0 , 18.5, 24.9, 29.9, float('inf')],
    labels =['Underweight', 'Healthy', 'Overweight', 'Obese']
    )
df_refined.head()
```

```python
df_refined = pd.get_dummies(df_refined, columns=['bmi_category'], dtype='int64', drop_first=-True)
df_refined.head()
```
